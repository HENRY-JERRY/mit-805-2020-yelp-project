```python
fileName = '../data/categories.txt'
category = [line.rstrip('\n') for line in open(fileName)]
category
```

```python
business_type = []
for i,j in zip(business['attributes'], business['categories']):
    if i:
        for k in i.keys():
            if k == 'BusinessAcceptsBitcoin':
                if i[k] == 'True' or i[k] == 'true':
                    if j:
                        word = j.split(',')
                        for w in word:
                            for c in category:
                                if w == c:
                                    business_type.append(c)
x = pd.Series(business_type)
x.value_counts()
```
